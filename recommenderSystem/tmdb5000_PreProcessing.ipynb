```python
import numpy as np
import pandas as pd
from utils import *
```

```python
movies_ = pd.read_csv("../unprocessedData/tmdb5000/tmdb_5000_movies.csv")
credits = pd.read_csv("../unprocessedData/tmdb5000/tmdb_5000_credits.csv")
```

```python
credits.rename(columns = {"movie_id" : "id"}, inplace = True)
```

```python
credits.head(1)
```

```python
movies = movies_.merge(credits, on = "id")
```

```python
movies.head()
```

```python
movies.columns
```

```python
movies = movies[["id", "original_title", "release_date", "genres", "keywords", "overview", "cast", "crew"]]
movies.head()
```

```python
movies.isna().sum()
```

```python
movies.shape
```

```python
movies.dropna(inplace = True)
movies.shape
```

```python
numberMoviesByYear = {}
for i in movies["release_date"]:
    if i[:4] not in numberMoviesByYear.keys():
        numberMoviesByYear[i[:4]] = 1
    else:
        numberMoviesByYear[i[:4]] += 1
```

```python
import collections
```

```python
numberMoviesByYearSorted = collections.OrderedDict(sorted(numberMoviesByYear.items()))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid")
```

```python
plt.figure(figsize = (15, 5))
plt.xticks(rotation = 90)
sns.barplot(x = list(numberMoviesByYearSorted.keys()), y = list(numberMoviesByYearSorted.values()))
```

```python
movies.drop(["release_date"], axis = 1,inplace = True)
```

```python
movies.head(1)
```

```python
movies["genres"] = movies["genres"].apply(convertGenres)
```

```python
movies["keywords"] = movies["keywords"].apply(convertGenres)
```

```python
movies["cast"] = movies["cast"].apply(convertCast)
```

```python
movies["crew"] = movies["crew"].apply(getDirector)
```

```python
movies["genres"] = movies["genres"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["keywords"] = movies["keywords"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["cast"] = movies["cast"].apply(lambda x:[i.replace(" ", "") for i in x])
movies["crew"] = movies["crew"].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies["overview"] = movies["overview"].apply(lambda x : x.split())
```

```python
movies.head()
```

```python
movies["Tags"] = movies["overview"] + movies["genres"] + movies["keywords"] + movies["cast"] + movies["crew"]
```

```python
movies["Tags"] = movies["Tags"].apply(lambda x : " ".join(x))
```

```python
data = movies[["id", "original_title", "Tags"]]
data.head()
```

```python
data["Tags"] = data["Tags"].apply(stem)
```

```python
data["Tags"] = data["Tags"].apply(lambda x : x.lower())
```
